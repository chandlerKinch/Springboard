```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
#Load dataset 
starcraft_data = pd.read_csv('wrangled_starcraft_data.csv')
starcraft_data.drop('Unnamed: 0', axis=1, inplace=True)
starcraft_data.head()
```

```python
#Scale data for K-nearest neighbors by min max scaling
minmax = MinMaxScaler()
#Not sure if we want to scale GameID and League Index or not. 
mm = minmax.fit_transform(starcraft_data.drop(['GameID', 'LeagueIndex'], axis=1))
mm_df = pd.DataFrame(mm, columns=starcraft_data.drop(['GameID', 'LeagueIndex'], axis=1).columns)
mm_df = starcraft_data[['LeagueIndex']].merge(mm_df, how='outer', left_index=True, right_index=True)
mm_df.head()
```

```python
#Scale data for Logisitc Regression
ss = StandardScaler()
scaler = ss.fit_transform(starcraft_data.drop(['GameID', 'LeagueIndex'], axis=1))
ss_df = pd.DataFrame(scaler, columns=starcraft_data.drop(['GameID', 'LeagueIndex'], axis=1).columns)
ss_df = starcraft_data[['LeagueIndex']].merge(ss_df, how='outer', left_index=True, right_index=True)
ss_df.head()
```

```python
#Create training and testing split for K neighbors 
X = mm_df.drop('LeagueIndex', axis=1)
y = mm_df['LeagueIndex']
X_train_K, X_test_K, y_train_K, y_test_K = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
#Create training and testing split for Logisitic Regression
X = ss_df.drop('LeagueIndex', axis=1)
y = ss_df['LeagueIndex']
X_train_L, X_test_L, y_train_L, y_test_L = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
#Create traing and testing split for Random Forest Tree
X = starcraft_data.drop('LeagueIndex', axis=1)
y = starcraft_data['LeagueIndex']
X_train_T, X_test_T, y_train_T, y_test_T = train_test_split(X, y, test_size=0.25, random_state=10)
```
